```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/match_ball_events.csv')
```

```python
matches = df['match_pkey'].unique()
len(matches)
```

```python
win_r=[]
win_w=[]
lose_r=[]
lose_w=[]

for match in matches:
    dfm = df[df['match_pkey']==match]
    pp1 = dfm[(dfm['over_num']<=6) & (dfm['innings']==1)]
    pp2 = dfm[(dfm['over_num']<=6) & (dfm['innings']==2)]
    ppr1 = sum(pp1['runs'])
    ppr2 = sum(pp2['runs'])
    ppw1 = sum(pp1['is_dismissal'])
    ppw2 = sum(pp2['is_dismissal'])
    team1 = sum(dfm[dfm['innings']==1]['runs'])
    team2 = sum(dfm[dfm['innings']==2]['runs'])
    
    if team1>team2:
        win_r.append(ppr1)
        win_w.append(ppw1)
        lose_r.append(ppr2)
        lose_w.append(ppw2)
        
    else:
        win_r.append(ppr2)
        win_w.append(ppw2)
        lose_r.append(ppr1)
        lose_w.append(ppw1)     
```

```python
sum(win_r)/len(win_r)
```

```python
sum(lose_r)/len(lose_r)
```

```python
sum(win_w)/len(win_w)
```

```python
pkeys = [11, 19, 42, 50, 61, 72]
players = ['Kohli', 'Faf', 'Watson', 'Padikkal', 'Utthapa', 'Warner']
av = []
sr = []
mat = []
for key in pkeys:
    df_p = df[(df['batsman_pkey']==key) & (df['over_num']<=6)]
    runs = sum(df_p['runs'])-sum(df_p['runs_wides'])-sum(df_p['runs_legbyes'])-sum(df_p['runs_noballs'])
    outs = sum(df_p['is_dismissal'])
    matches = len(df_p['match_pkey'].unique())
    mat.append(matches)
    noballs = len(df_p[df_p['runs_noballs']!=0])
    wides = len(df_p[df_p['runs_wides']!=0])
    legal_balls = len(df_p) - noballs-wides
    av.append(runs/outs)
    sr.append(runs/legal_balls*100)
```

```python
print(mat)
```

```python
# creating the bar plot
plt.bar(players, av, color ='maroon', width = 0.4)
 
plt.xlabel("Batsmen")
plt.ylabel("Batting Average")
plt.title("Batting averages in Powerplay")
plt.show()
```

```python
X_axis = np.arange(len(players))
plt.bar(X_axis - 0.2, av, 0.4, label = 'Average')
plt.bar(X_axis + 0.2, sr, 0.4, label = 'Strike-Rate')
  
plt.xticks(X_axis, players)
plt.xlabel("Batsmen")
plt.title("Batting Performance in Powerplays")
plt.legend()
plt.savefig('bat_pp.png')
```
