```python
import pandas as pd
import numpy as np
from utils import get_BP, get_CBR
```

```python
df = pd.read_csv('data/auction_stats.csv')
df
```

```python
df_final = pd.DataFrame()
for i in range(len(df)):
    temp = dict(df.iloc[i])
    temp['bat_powerplay'] = get_BP(temp['player_pkey'], 'powerplay')
    temp['bat_death'] = get_BP(temp['player_pkey'], 'death')
    temp['bowl_powerplay'] = get_CBR(temp['player_pkey'], 'powerplay')
    temp['bowl_death'] = get_CBR(temp['player_pkey'], 'death')
    
    df_final = df_final.append(temp, ignore_index=True)
```

```python
df_final
```

```python
df_final.to_csv('data/auction_stats_special.csv')
```
